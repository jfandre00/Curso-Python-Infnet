#1. Qual a diferença entre um DataFrame e uma Series no Panda? Justifique sua resposta

No Pandas, DataFrame e Series são duas estruturas de dados fundamentais, mas possuem diferenças importantes.

a.	Series

    i.	É uma matriz unidimensional rotulada, capaz de armazenar qualquer tipo de dados (inteiro, string, float, etc).

    ii.	Cada elemento em uma Series tem um índice associado.

    iii.	É semelhante a uma coluna em uma tabela ou uma lista com um índice.

b.	DataFrame

    i.	É uma estrutura de dados bidimensional – uma tabela de dados com linhas e colunas.

    ii.	Cada columa de um DataFrame é uma Series

    iii.	Pode ser visto como um dicionário de objetos Series alinhados em colunas.

    iv.	É ideal para manipular e analisar planilhas ou tabelas em banco de dados.


#2. Quais tipos de dados o Pandas aceita e como ele os armazena internamente? Justifique sua resposta com exemplos concretos.

O Pandas é uma biblioteca muito versátil que aceita diversos tipos de dados e os armazena de maneira 
eficiente. Os principais tipos de dados que o Pandas aceita e como ele os armazena internamente são:

    a.	Numéricos (inteiros (int64) e floats (float64))
Internamente, o Pandas utiliza arrays do NumPy para armazenar esses tipos de dados eficientemente.

**Exemplo: data = {‘Inteiros’: [1, 2, 3], ‘Floats’: [1.1, 2.2, 3.3]}**

    b.	Texto (strings (object))
As strings são armazenadas como objetos (object) e manipuladas eficientemente para operações de texto.

**Exemplo: data = {‘Texto’: [‘A’, ‘B’, ‘C’]}**

    c.	Datas e tempos (datetime (datetime64))
O Pandas tem suporte nativo para manipular datas e tempos, utilizando arrays de datetime64 do NumPy.

**Exemplo: data = {‘Datas’: pd.date_range(‘20250101’, periods=3)]**

    d.	Dados categóricos (categorical)
O Pandas tem um tipo de dado especial para representar dados categóricos, o que melhora a eficiência de armazenamento e a velocidade de operações.

**Exemplo: data = {‘Categoria’: pd.Categorical([‘A’, ‘B’, ‘A’])}**

    e.	Dados booleanos (bool)
Internamente, o Pandas utiliza arrays de booleanos para armazenar esses tipos de dados.

**Exemplo: data = {‘Booleano’: [True, False, True]}**

    f.	Dados ausentes (NaN – Not a Number)
O Pandas utiliza o NaN para representar dados ausentes em arrays de ponto flutuante.

**Exemplo: data = {‘Dados’: [1, 2, np.nan]}**


#3. De que é composto um DataFrame? Explique as partes principais (colunas, índices, valores, tipos de dados).

Um DataFrame no Pandas é uma estrutura de dados bidimensional semelhante a uma tabela. As suas principais partes são:

    a.	Colunas 

        i.	Cada coluna em um DataFrame representa uma variável e é armazenada como uma Series.
        ii.	As colunas têm nomes, que podem ser strings ou outros tipos de dados.

    b.	Índices

        i.	O índice é a etiqueta associada a cada linha de dados.
        ii.	Por padrão, o índice é uma sequência numérica iniciada em zero, mas pode ser configurado para representar outras informações, como datas ou identificadores únicos. 

    c.	Valores

        i.	Os valores são os dados armazenados no DataFrame, organizados em linhas e colunas.
        ii.	Cada célula na tabela representa um valor único na interseção de uma linha e uma coluna.
        
    d.	Tipos de dados
    
        i.	Cada coluna em um DataFrame pode ter um tipo de dado específico, como numérico, texto, datas, booleanos, etc.
        ii.	O Pandas garante eficiência na manipulação e armazenamento desses tipos de dados utilizando arrays do NumPy, conforme já explicado acima.

```python
#7. Quantas categorias diferentes existem na coluna 'workclass'?

df['workclass'].groupby(df['workclass']).count()

#Existem 9 categorias diferentes na coluna 'workclass', considerando que a categoria '?' também é uma categoria.
```

```python
#8. Quantas pessoas do dataset possuem nível de educação de Bachelors?

df['education'].groupby(df['education'] == 'Bachelors').count()

#Existem 8025 pessoas no dataset que possuem nível de educação de Bachelors.
```

```python
#9. Qual é a média de horas trabalhadas por semana (hours-per-week) para quem ganha mais de 50K e para quem ganha menos de 50K?

print(df['hours-per-week'].groupby(df['income'] == '>50K').mean())

#A média de horas trabalhadas por semana para quem ganha mais de 50K é de 45.47 horas e para quem ganha menos de 50K é de 38.84 horas.

```

```python
#10. Quais são as 3 ocupações (ocupation) mais comuns entre pessoas que ganham mais de 50K?


novo_df = df['occupation'].groupby(df['income'] == '>50K').value_counts()
novo_df[True].head(3)

#As 3 ocupações mais comuns entre pessoas que ganham mais de 50K são: Prof-specialty, Exec-managerial e Sales.
```

```python
#14. Crie uma nova coluna capital_difference, que é a diferença entre capital-loss e capital-gain.

df['capital_difference'] = df['capital-loss'] - df['capital-gain']
df['capital_difference']

#A nova coluna capital_difference foi criada com sucesso, e a diferença entre capital-loss e capital-gain foi calculada.
```

```python
#15. Crie uma nova coluna age_decade que indica em qual década a pessoa nasceu (Ex. se a idade é 37, então age_decade = 30s).

df['age_decade'] = (df['age'] // 10 * 10).astype(str) + 's'
df['age_decade']

#Para resolver a questão, utilizei a função .astype(str) para converter a coluna 'age_decade' para string e adicionei a string 's' ao final de cada valor da coluna 'age_decade'.
```

```python
#16. Crie uma nova coluna marital_status_code que converte marital_status em variáveis numéricas (0, 1, 2, ...) para cada classe diferente.

df['marital_status_code'] = df['marital-status'].astype('category').cat.codes
df['marital_status_code']

#Para resolver a questão, utilizei a função .astype('category').cat.codes para converter a coluna 'marital-status' em variáveis numéricas. Os valores numéricos atribuídos foram:
# 0 para Divorced
# 1 para Married-AF-spouse
# 2 para Married-civ-spouse
# 3 para Married-spouse-absent
# 4 para Never-married
# 5 para Separated
# 6 para Widowed
```
