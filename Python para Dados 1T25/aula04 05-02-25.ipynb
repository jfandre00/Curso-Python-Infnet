```python
#Exercício 4 - Agrupando os pinguins por espécie e calculando a média do peso corporal

df['body_mass_g'].groupby(df['species']).mean()
```

```python
#Exercício 5 - Criando uma nova coluna com a razão entre o comprimento da nadadeira e o peso corporal
df['flipper_ratio'] = df['flipper_length_mm'] / df['body_mass_g']
df.head()
```
